#DISTRIBUCION LOGNORMAL - TIEMPOS DE INACTIVIDAD
La distribución Lognormal es usada principalmente para modelar tiempos de servicio o inactividad que tienen una asimetría positiva, es decir, donde hay más eventos de corta duración, pero algunos pueden ser mucho más largos.
##Aplicacion
En una línea de producción automatizada en la fabricación de chips electrónicos, existen brazos automatizados que se encargan de ensamblar y soldar componentes en las placas de circuitos integrados.
En esta parte del proceso de fabricación, es común que los equipos automatizados sufran interrupciones debido a mantenimiento, fallas técnicas menores, o ajustes en los materiales.
Estas pausas tienden a seguir una distribución lognormal: muchas son breves, pero ocasionalmente pueden ser más prolongadas debido a problemas más complejos.
El uso del método de la transformada inversa permite simular estos tiempos de inactividad, ayudando a evaluar la eficiencia del sistema y a planificar el mantenimiento preventivo para minimizar las interrupciones en la producción de chips.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución lognormal
mu = 1.0   # Media de la variable logarítmica
sigma = 0.5  # Desviación estándar de la variable logarítmica
n = 100  # Número de variables aleatorias a generar

# Generar n números aleatorios U ~ Uniform(0, 1)
U = np.random.uniform(0, 1, n)

# Transformada inversa para generar tiempos de inactividad lognormales
Z = mu + sigma * stats.norm.ppf(U)
T = np.exp(Z)  # Tiempos generados

# Mostrar los resultados
print("Tiempos de inactividad generados:", T)

# Graficar histograma de los tiempos de inactividad generados
plt.hist(T, bins=20, edgecolor='black', density=True)

# Graficar la función teórica de densidad de la distribución lognormal
x = np.linspace(min(T), max(T), 100)
pdf = stats.lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf, 'r', label='Distribución lognormal teórica')

plt.title("Histograma de los tiempos de inactividad (Distribución Lognormal)")
plt.xlabel("Tiempo de inactividad")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.show()
```

#CONCLUSION
La simulación de los tiempos de inactividad bajo una distribución Lognormal nos proporciona un entendimiento profundo de la variabilidad y la complejidad de los eventos de pausa en sistemas automatizados. Algunos puntos clave para destacar son:

Interpretación de los tiempos de inactividad: Los tiempos generados muestran una mayor frecuencia de pausas breves y algunas interrupciones más largas, lo cual es característico de la distribución lognormal. Este comportamiento refleja escenarios donde la mayoría de las interrupciones son de corta duración (fallos menores o ajustes rápidos), pero existe la posibilidad de pausas más extensas debido a problemas complejos.

Asimetría positiva: La distribución lognormal, al tener una asimetría positiva, permite modelar adecuadamente situaciones donde los tiempos de inactividad son mayormente cortos, pero con eventos ocasionales y más largos que pueden impactar significativamente el rendimiento del sistema.

Impacto operativo: Entender y modelar estos tiempos de inactividad ayuda a los ingenieros de producción y mantenimiento a tomar decisiones más informadas para minimizar tiempos muertos. Identificar las causas principales de los tiempos más prolongados permite establecer estrategias preventivas más efectivas y mejorar la planificación del mantenimiento correctivo y preventivo.

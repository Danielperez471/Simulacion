#DISTRIBUCION EXPONENCIAL - TIEMPO DE SERVICIO.

###La distribución exponencial es una distribución de probabilidad continua que se utiliza para modelar el tiempo entre eventos, donde los eventos ocurren de manera independiente a una tasa constante.

###Facilitador: José Gabriel Rodríguez Rivas

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de servicio)
lambda_servicio = 1 / 2  # Tiempo promedio de servicio = 2 minutos

# Función para generar un tiempo de servicio
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_servicio

# Generar 100 tiempos de servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(100)]

# Mostrar los tiempos de servicio
print("Tiempos de servicio generados:", tiempos_servicio)

# Graficar los tiempos de servicio simulados
plt.hist(tiempos_servicio, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

#CONCLUSION
al observar los datos de haber generado estas 100 variables aleatorias, se detecto lo siguiente:

Distribución de los datos: Los tiempos de servicio oscilan desde valores muy cercanos a cero hasta tiempos más largos cercanos a los 9 minutos. Esto es consistente con la naturaleza de la distribución exponencial, donde la mayoría de los eventos ocurren en tiempos cortos, pero hay una baja probabilidad de tiempos más largos.

Tiempo promedio: Como se utilizó un parámetro de tasa 𝜆=1/2, el tiempo promedio esperado de servicio es de 2 minutos. Esto se refleja aproximadamente en los datos simulados, aunque pueden aparecer desviaciones debido a la aleatoriedad de la simulación.

Simulación realista: La variabilidad observada en los tiempos de servicio refleja adecuadamente situaciones del mundo real, donde algunos clientes pueden ser atendidos rápidamente mientras que otros requieren más tiempo.

Usos prácticos: Estos resultados permiten analizar y optimizar el funcionamiento del sistema de atención. Por ejemplo, al conocer los tiempos de servicio típicos y los más largos, se pueden tomar decisiones para mejorar la eficiencia del personal o ajustar los recursos disponibles durante horas pico.

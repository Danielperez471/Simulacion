#DISTRIBUCION EXPONENCIAL - TIEMPO DE SERVICIO.

###La distribuci칩n exponencial es una distribuci칩n de probabilidad continua que se utiliza para modelar el tiempo entre eventos, donde los eventos ocurren de manera independiente a una tasa constante.

```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metro de la distribuci칩n exponencial (tasa de servicio)
lambda_servicio = 1 / 2  # Tiempo promedio de servicio = 2 minutos

# Funci칩n para generar un tiempo de servicio
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)  # Generar n칰mero aleatorio uniforme
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_servicio

# Generar 100 tiempos de servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(100)]

# Mostrar los tiempos de servicio
print("Tiempos de servicio generados:", tiempos_servicio)

# Graficar los tiempos de servicio simulados
plt.hist(tiempos_servicio, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribuci칩n de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

#CONCLUSION
al observar los datos de haber generado estas 100 variables aleatorias, se detecto lo siguiente:

Distribuci칩n de los datos: Los tiempos de servicio oscilan desde valores muy cercanos a cero hasta tiempos m치s largos cercanos a los 9 minutos. Esto es consistente con la naturaleza de la distribuci칩n exponencial, donde la mayor칤a de los eventos ocurren en tiempos cortos, pero hay una baja probabilidad de tiempos m치s largos.

Tiempo promedio: Como se utiliz칩 un par치metro de tasa 洧랝=1/2, el tiempo promedio esperado de servicio es de 2 minutos. Esto se refleja aproximadamente en los datos simulados, aunque pueden aparecer desviaciones debido a la aleatoriedad de la simulaci칩n.

Simulaci칩n realista: La variabilidad observada en los tiempos de servicio refleja adecuadamente situaciones del mundo real, donde algunos clientes pueden ser atendidos r치pidamente mientras que otros requieren m치s tiempo.
